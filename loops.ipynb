Assignment 1: The Multiplication Table (for loop)
The Goal: Write a program that asks the user for a number, and then prints out the multiplication table for that number from 1 to 10.


```python
number=int(input())
for i in range(1,10):
    x= number*i
    print("{} x {} = {}".format(number,i,x))
```

Assignment 2: The Vowel Counter (for loop)
The Goal: Write a program that asks the user to type in a word or sentence. The program will count exactly how many vowels (a, e, i, o, u) are in that string and print the final count.


```python
stringname="New course is the Mastery in Python Programming"
count=0
stringname=stringname[0:]
vowels=['a','e','i','o','u']
for j in range(0,len(vowels)):
    for i in range(0,len(stringname)):
        if stringname[i]==vowels[j]:
            count=count+1
print(count)

#Optimized
stringname = input("Enter a word or sentence: ").lower()
count = 0
vowels = ['a','e','i','o','u']
for char in stringname:
    if char in vowels:
        count += 1
print(f"Number of vowels: {count}")
```

Assignment 3: The "Higher or Lower" Guessing Game (while loop)
The Goal: The computer has a secret number (you can hardcode this, like secret = 9). The user has to guess the number. If they guess wrong, tell them if they need to guess higher or lower. Keep asking them until they get it right. Include a rule to check if their guess is outside a valid range (e.g., above 100).


```python
#For
secret=92
for i in range(0,1000):

    inputnumber = int(input("Enter the new guess number"))
    if inputnumber < 1 or inputnumber > 100:
        print("Please guess a number between 1 and 100!")
        continue

    if inputnumber >secret:
        print("Secret is lesser than {}".format(inputnumber ))
    elif inputnumber < secret:
        print("Secret is greater than {}".format(inputnumber ))
    elif inputnumber ==secret:
        print("You found the number correct {}.".format(inputnumber))
        break

#While(Thats why we are using while no need to write range ahead)
secret = 92
while True:
    inputnumber = int(input("Enter your guess (1-100): "))
    
    if inputnumber < 1 or inputnumber > 100:
        print("Please guess a number between 1 and 100!")
        continue
    
    if inputnumber > secret:
        print("Secret is lesser than {}".format(inputnumber))
    elif inputnumber < secret:
        print("Secret is greater than {}".format(inputnumber))
    elif inputnumber == secret:
        print("You found the number correct {}!".format(inputnumber))
        break
```

Assignment 4: The ATM Lockout (while loop)
The Goal: Recreate an ATM PIN login. The user has a correct PIN (e.g., account_pin = 12345). They are allowed a maximum of 3 attempts to get it right. If they get it wrong 3 times, the loop stops and tells them their account is locked. If they get it right, print a success message and stop the loop.

```python
account_pin = 12345
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    guess = int(input("Enter your PIN: "))
    attempts += 1
    if guess == account_pin:
        print("PIN accepted. Welcome!")
        break
    else:
        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"Incorrect PIN. You have {remaining} attempt(s) left.")
        else:
            print("Account locked. Too many incorrect attempts.")
```
