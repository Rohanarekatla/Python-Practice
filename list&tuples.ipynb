Question: Using list comprehension, how can you generate a list of cubes for all odd numbers within a specified range (e.g., up to 20)?

Test Case Input: range(10) (for odd numbers)

Test Case Output: [1, 27, 125, 343, 729]



```python
numbers=[i**3 for i in range(1,11) if i%2==1]
print(numbers)
```


Question: How can you create a list of lists, where each inner list contains the index, the word, and the length of the word, from a given sentence, using list comprehension and enumerate?

Test Case Input: sentence = "Python is fun"

Test Case Output: [[0, 'Python', 6], [1, 'is', 2], [2, 'fun', 3]]



```python
inputlistwords="Python is fun"
list2 = [[i, word, len(word)] for i, word in enumerate(inputlistwords.split())]
print(list2)
```

Test Case Input: celsius_temps = [0, 25]

Test Case Output: [32.0, 77.0]

Question: How can you filter a list of strings to include only those that contain a specific character (case-insensitively), using list comprehension?

Test Case Input: fruits = ["grape", "orange", "kiwi", "pear"], search for 'e'

Test Case Output: ['grape', 'orange', 'pear']




```python
fruits = ["grape", "orange", "kiwi", "pear"]
char = 'e'
output = [e for e in fruits if char in e.lower()]
print(output)
```

Question: How can you flatten a list of lists (matrix) into a single-dimensional list using nested list comprehension?

Test Case Input: matrix = [[1, 2], [3, 4]]

Test Case Output: [1, 2, 3, 4]



```python
matrix = [[1, 2], [3, 4]]
flatten = [g for row in matrix for g in row]
print(flatten)
```


Question: How can you process a list containing mixed data types, including numbers (as integers and strings) and non-numeric strings, to extract only the valid numeric entries and convert them all into integers?

Test Case Input: mixed_data = [5, "abc", "15", 2.0, "error", 30]

Test Case Output: [5, 15, 2, 30]

```python
mixed_data = [5, "abc", "15", 2.0, "error", 30]
valid = [int(item) for item in mixed_data if isinstance(item, (int, float)) or (isinstance(item, str) and item.isdigit())
]
print(valid)
```

Level 2: Intermediate (Immutability & Methods)
The Workaround: You are given a tuple of colors. You realize you need to replace "blue" with "purple". Since tuples are immutable, you cannot simply do colors[2] = "purple". Write the code to successfully update the collection of colors to include "purple" instead of "blue", ensuring your final colors variable is still a tuple.

Test Case:

Input: colors = ("red", "green", "blue", "yellow")

Expected Final colors Tuple: ("red", "green", "purple", "yellow")



```python
colors = ("red", "green", "blue", "yellow")
colors_list=list(colors)
blue_index=colors_list.index("blue")
colors_list[blue_index]="purple"
colors=tuple(colors_list)
print(colors)
```

The Roster Sort: You have a list of tuples where each tuple contains a student's name and their final score. Write code to sort this list from the highest score to the lowest score using a lambda function.

Test Case:

Input: students = [("Alice", 88), ("Bob", 95), ("Charlie", 72), ("Diana", 90)]

Expected Output: [('Bob', 95), ('Diana', 90), ('Alice', 88), ('Charlie', 72)]

```python
students = [("Alice", 88), ("Bob", 95), ("Charlie", 72), ("Diana", 90)]
def get_score(student):
    return student[1]

students.sort(key=get_score, reverse=True)
print(students)
```
