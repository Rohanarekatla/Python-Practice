1. The Deduplicator (Data Cleaning)
You are processing a log of user IDs that visited your website today, but many users refreshed the page and appear multiple times.
​Given: daily_visitors = [101, 105, 102, 101, 103, 105, 108, 102]
​Task: Write a script that removes all duplicates and prints a standard Python list of the unique visitor IDs, sorted in ascending order.

2. The Mutual Friends (Intersection)
You are building a social media feature that shows two users who their mutual friends are.
​Given: user1_friends = ["Alice", "Bob", "Charlie", "David"]
user2_friends = ["Bob", "David", "Eve", "Frank"]
​Task: Convert these lists to sets and use a single set operator to find the friends they have in common.


````python
user1_friends = ["Alice", "Bob", "Charlie", "David"]
user1_friends = set(user1_friends)
user2_friends = ["Bob", "David", "Eve", "Frank"]
user2_friends = set(user2_friends)
print(user1_friends.intersection(user2_friends))
````

3. The Missing Ingredients (Difference)

4. The Vowel Checker (Subsets)
Sets are incredibly fast for checking if one group of items is entirely contained within another.
​Given: vowels = {"a", "e", "i", "o", "u"}
​Task: Write a function that takes a word as input (e.g., "sequoia" or "apple"). Convert the input word into a set of its characters, and use a set method to check if the word contains all five vowels (i.e., check if vowels is a subset of the word's characters).


````python
vowels = {"a", "e", "i", "o", "u"}
set(vowels)
word="sequoia"
new=set(list(word))
print(new.issuperset(vowels))
````

5. The Exclusive Club (Symmetric Difference)
You are organizing a team-building event and have two lists of employees who signed up for either Paintball or Go-Karting. You want to find the people who signed up for only one activity, excluding the eager people who signed up for both.
​Given:
paintball = {"John", "Sarah", "Mike", "Emma"}
gokarting = {"Sarah", "Emma", "David", "Tom"}

````python
paintball = {"John", "Sarah", "Mike", "Emma"}
gokarting = {"Sarah", "Emma", "David", "Tom"}
paintball.symmetric_difference(gokarting ) 
````
